```python
import geopandas as gpd
import pandas as pd
```

```python
df = pd.read_csv('/data/2022/Zensus2022_Durchschn_Nettokaltmiete_100m-Gitter.csv',sep=';',decimal=',',index_col=0)
```

```python
df
```

```python
df['x_mp_100m']
```

```python
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df['x_mp_100m'], df['y_mp_100m']), crs='EPSG:3035')
```

```python
gdf
```

```python
# Buffer the points using a square cap style
# Note cap_style: round = 1, flat = 2, square = 3
gdf['geometry'] = gdf['geometry'].buffer(50, cap_style = 3)
```

```python
gdf.cx[4150000:4170000,2760000:2780000].explore(column='durchschnMieteQM',cmap='coolwarm')
```
