```python
# Load data
employment = pd.read_excel('/data/uscuni-restricted/Data_SLDB_2021/zsj_data/ZSJD_slouceneZSJD/sloucene/nadzsjd_zamestnani_2021.xlsx')
```

```python
employment = employment.iloc[5:]
employment.head(2)
```

```python
employment = employment.iloc[:,12:]
employment
```

```python
employment = employment.rename(columns={
    'nadzsjd':'kod_nadzsj_d'
})
```

```python
employment
```

```python
prevodnik.head(10)
```

```python
prevodnik['kod_nadzsj_d'].duplicated()
```

```python
nadzsj_emp = pd.merge(prevodnik, employment,left_on = 'kod_nadzsj_d',right_on ='kod_nadzsj_d',how = 'left')
nadzsj_emp.head(10)
```

```python
nadzsj_emp.
```

```python
grouped = nadzsj_emp.groupby('kod_nadzsj_d',as_index=False).agg('first')
```

```python
grouped
```

```python
pol = polygons.iloc[:,[0, 1, -1]]
pol
```

```python
nadzsj_emp_pol = pd.merge(grouped, pol, left_on = 'kod_zsj', right_on = 'KOD_ZSJ_P', how = 'inner')
nadzsj_emp_pol = nadzsj_emp_pol.drop(columns={'KOD_ZSJ_P','NAZ_ZSJ_P'})
nadzsj_emp_pol
```

```python
pol.crs
```

```python
gdf = gpd.GeoDataFrame(nadzsj_emp_pol, geometry='geometry').set_crs(pol.crs)
```
