```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
```

```python
polygons = gpd.read_file('/data/uscuni-restricted/Data_SLDB_2021/zsj_sldb_2021_-7697413790081074647.zip')
```

```python
polygons.head(2)
```

```python
prevodnik = pd.read_excel('/data/uscuni-restricted/Data_SLDB_2021/zsj_data/ZSJD_slouceneZSJD/sloucene/_prevodnik_zsjd_nadzsjd.xlsx')
```

```python
prevodnik
```

```python
prevodnik.columns = prevodnik.iloc[2]
prevodnik = prevodnik.iloc[4:]
prevodnik = prevodnik.iloc[:,18:]
```

```python
prevodnik
```

```python
pol = polygons.iloc[:,[0, 1, -1]]
pol
```

```python
pp = pd.merge(prevodnik, pol, left_on = 'kod_zsj', right_on = 'KOD_ZSJ_P')
pp
```

```python
pp[pp.geometry.duplicated(False)].sort_values('kod_zsj')
```

```python

filtered = pp[pp['slouceni'] == 'ANO']

# Group the filtered DataFrame and apply the aggregation
grp = filtered.groupby('kod_nadzsj_d', as_index=False).agg('first')
```

```python
# First, filter the rows where 'merged' is 'yes' and group them by 'kod_nadzsj_d'
aggregated = pp[pp['slouceni'] == 'ANO'].groupby('kod_nadzsj_d', as_index=False).agg('first')

# Now, get the rows where 'merged' is not 'yes' (these should remain unchanged)
non_merged = pp[pp['slouceni'] != 'ANO']

# Concatenate the two DataFrames: aggregated and non_merged
result = pd.concat([aggregated, non_merged]).sort_index()
result
```

```python
res = result.drop(columns = "geometry")
```

```python
res
```

```python
a = pd.merge(res, pol, left_on='kod_zsj', right_on='KOD_ZSJ_P',how = 'right')
a
```

```python
gdf = gpd.GeoDataFrame(a, geometry='geometry')
```

```python
gdf.plot()
```

```python
dissolved_gdf = gdf.dissolve(by='kod_nadzsj_d', aggfunc='first')
```

```python
dissolved_gdf.plot()
```
