```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
```

```python
polygons = gpd.read_file('/data/uscuni-restricted/Data_SLDB_2021/zsj_sldb_2021_-7697413790081074647.zip')
```

```python
polygons.head(2)
```

```python
prevodnik = pd.read_excel('/data/uscuni-restricted/Data_SLDB_2021/zsj_data/ZSJD_slouceneZSJD/sloucene/_prevodnik_zsjd_nadzsjd.xlsx')
```

```python
prevodnik.head(5)
```

```python
prevodnik.columns = prevodnik.iloc[2]
prevodnik = prevodnik.iloc[4:]
prevodnik = prevodnik.iloc[:,18:]
```

```python
pol = polygons.iloc[:,[0, 1, -1]]
pol = pol.rename(columns={'KOD_ZSJ_P':'kod_zsj'})
pol.head(2)
```

```python
# Load data
employment = pd.read_excel('/data/uscuni-restricted/Data_SLDB_2021/zsj_data/ZSJD_slouceneZSJD/sloucene/nadzsjd_zamestnani_2021.xlsx')
```

```python
employment.head(5)
```

```python
employment.iloc[3, 0:13] = employment.iloc[4, 0:13]
```

```python
employment.columns = employment.iloc[3]
employment.head(5)
```

```python
employment = employment.iloc[5:]
employment.head(2)
```

```python
employment = employment.iloc[:,12:]
employment
```

```python
employment = employment.rename(columns={
    'nadzsjd':'kod_nadzsj_d'
})
```

```python
prevodnik.head(2)
```

```python
employment.head(2)
```

```python
emp_p = pd.merge(employment, prevodnik[['kod_zsj', 'kod_nadzsj_d']], on='kod_nadzsj_d', how='left')
```

```python
emp_p.head(2)
```

```python
pol.head(2)
```

```python
emp_df = pd.merge(emp_p, pol, on='kod_zsj', how='left')
```

```python
emp_gdf = gpd.GeoDataFrame(emp_df, geometry='geometry', crs=pol.crs)
```

```python
emp_dissolved = emp_gdf.dissolve(by='kod_nadzsj_d', aggfunc='first', as_index=False)
```

```python
emp_dissolved.head(2)
```
